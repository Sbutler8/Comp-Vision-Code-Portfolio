<h2><b>[Part 3] Calculate IOU (Intersection over Union)</b></h2>
<h5>In Python, given any two lists that defined the rectangles of the minimum and maximum corners: </h5>

```python
img = np.zeros((20,20,3), np.uint8)

ground_truth = [(3,6),(10,10)]
prediction = [(7,-1),(15,7)]

cv2.rectangle(img,ground_truth[0],ground_truth[1],(255,0,0),1)
cv2.rectangle(img,prediction[0],prediction[1],(0,255,0),1)
# plt.imshow(img)

def IOU(groundTruthBox, predictionBox):
    # RectAreas
    groundTruthArea = (groundTruthBox[1][1] - groundTruthBox[0][1]) * (groundTruthBox[1][0] - groundTruthBox[0][0])
    predictArea = (predictionBox[1][1] - predictionBox[0][1]) * (predictionBox[1][0] - predictionBox[0][0])
    print('Ground Truth Box Area: ' + str(groundTruthArea))
    print('Prediction Box Area: ' + str(predictArea))
    
    #XY coordinates of intersection
    groundTruthBox_X = max(groundTruthBox[0][0], predictionBox[0][0])
    groundTruthBox_Y = max(groundTruthBox[0][1], predictionBox[0][1])
    predictionBox_X = min(groundTruthBox[1][0], predictionBox[1][0])
    predictionBox_Y = min(groundTruthBox[1][1], predictionBox[1][1])
    
    # Intersection Area
    interArea = max(0,predictionBox_X - groundTruthBox_X) * max(0,predictionBox_Y - groundTruthBox_Y)
    print ('Intersection Area: ' + str(interArea))
    
    return interArea / (groundTruthArea + predictArea - interArea)


print ('IOU: ' + str(IOU(ground_truth, prediction)))
```

```python
from matplotlib.ticker import MaxNLocator
import itertools

def draw_rectangles(rectangleDataList):
    padding = 5
    maxX = max([max_x for (min_x,min_y,max_x,max_y) in rectangleDataList]) + padding
    minX = min([min_x for (min_x,min_y,max_x,max_y) in rectangleDataList]) - padding
    maxY = max([max_y for (min_x,min_y,max_x,max_y) in rectangleDataList]) + padding
    minY = min([min_y for (min_x,min_y,max_x,max_y) in rectangleDataList]) - padding

    # this plots all of the rectangles on the same graph
    fig, ax = plt.subplots()
    plt.title('Rectangles')
    ax.set_aspect(1) #normalizes the graph
    ax.yaxis.set_major_locator(MaxNLocator(integer=True))
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))

    #the x and y limits for the graph are set by the largest x and y values generated and are restricted by radii
    plt.xlim(minX , maxX )
    plt.ylim(minY , maxY )
    plt.grid(True, which='both')

    for rectangle in rectangleDataList:
        min_x,min_y,max_x,max_y = rectangle
        width = max_x - min_x
        height = max_y - min_y

        # Plot library uses min left as the point
        pt_x = min_x
        pt_y = min_y

        # For annotation
        center_x = min_x + width/2.0
        center_y = min_y + height/2.0

        # Place the rectangle
        rectangleObj = plt.Rectangle(xy=(pt_x, pt_y), width=width, height=height, color='b', fill=False, linewidth=2)
        ax.add_artist(rectangleObj)
        annotate_string = 'Min: (' + str(min_x)+','+str(min_y)+')\n'+'Max: ('+str(max_x)+','+str(max_y) + ')'
        label = ax.annotate(annotate_string, xy=(center_x, center_y), fontsize=9, ha="center")

    plt.show()

draw_rectangles([list(itertools.chain(*ground_truth)), list(itertools.chain(*prediction))])
```

<h2>AdaBoost With Code</h2>

```python
from collections import defaultdict 
boost_predictor_class = defaultdict(dict)
boost_predictor_class['A'] = {'blue':[5,6,7,8,9], 'red':[1,2,3,4]}
boost_predictor_class['B'] = {'blue':[6,7,8,9], 'red':[1,2,3,4,5]}
boost_predictor_class['C'] = {'blue':[3,4,5,9], 'red':[1,2,6,7,8]}
boost_predictor_class['D'] = {'blue':[1,6,7,8,9], 'red':[2,3,4,5]}

ground_truth = {'blue':[2,4,7,8,9], 'red':[1,3,5,6]}

error_assignment = {point:1 for point in range(1,10)}
```

```python
print (boost_predictor_class)

print (ground_truth)

print (error_assignment)
```

```python
def find_best_predictor(boost_predictor_class, ground_truth, error_assignment, classes):
    score_list = list()
    for predictor in boost_predictor_class:
        error_list = [set(ground_truth[class_type]) - set(boost_predictor_class[predictor][class_type]) for class_type in classes] 
        error_list = list(itertools.chain(*error_list))
        weighted_error = sum([error_assignment[item] for item in error_list])
        print (predictor, error_list, weighted_error) 
        score_list.append((weighted_error,predictor, error_list))
    print (score_list)
    
    candidate = min(score_list, key=lambda t: (t[0], len(t[2])))
    return (candidate[1],candidate[2])    
```

```python
find_best_predictor(boost_predictor_class, ground_truth, error_assignment, ['blue','red'])
```

```python
for iteration in range (len(boost_predictor_class)):
   
   # find best predictor and make list of errors 
   best_predictor, error_list = find_best_predictor(boost_predictor_class, ground_truth, error_assignment, ['blue','red'])

   print ('Iteration ', iteration, ' Best Predictor ', best_predictor, ' error list ', error_list)
   # assign new weight to error dictionary
   [error_assignment.update({key: error_assignment[key] * 2}) for key in error_list]

   print ('new error weights', error_assignment)
   # remove best predictor from consideration
   del boost_predictor_class[best_predictor]
```
